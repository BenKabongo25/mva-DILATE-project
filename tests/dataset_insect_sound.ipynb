```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from scipy.io import arff
plt.style.use("bmh")
```

```python
DATA_PATH = "../data/InsectSound/"
data_train = pd.DataFrame(arff.loadarff(DATA_PATH + "InsectSound_TRAIN.arff")[0])
data_test = pd.DataFrame(arff.loadarff(DATA_PATH + "InsectSound_TEST.arff")[0])
data_train.head()
```

```python
X_train = np.array(data_train.drop("target", axis=1), dtype=float)[:, 200:500]
y_train = np.array(data_train["target"])

X_test = np.array(data_test.drop("target", axis=1) , dtype=float)[:, 200:500]
y_test = np.array(data_test["target"])

X_train.shape, X_test.shape
```

```python
for y in np.unique(y_train):
    plt.figure(figsize=(30, 5))
    print(y)
    Xy_train = X_train[y_train == y]
    for i, idx in enumerate(np.random.randint(len(Xy_train), size=5), 1):
        plt.subplot(1, 5, i)
        plt.plot(np.arange(200, 500), Xy_train[idx])
    plt.show()
```
